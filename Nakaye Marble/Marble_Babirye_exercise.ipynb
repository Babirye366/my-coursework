```python
#1.	Write a Pandas program to compare the elements of two Pandas series. 
#Sample Series [4,65,436,3,9], [7,0,3,897,9]
# Assume the series are called ds1 and ds2 respectively
import pandas as pd
ds1 = pd.Series([4,65,436,3,9])
ds2 = pd.Series([7,0,3,897,9])
print("Compare the elements of the said Series:")
print("Equals:")
print(ds1 == ds2)
print("Greater than:")
print(ds1 > ds2)
print("Less than:")
print(ds1 < ds2)
```

```python
# 2.2.	Write a Pandas program to add, subtract, multiply and divide two Pandas series. 
#Sample Series: [2,4,6,8,14], [1,3,5,7,9]

ds1 = pd.Series([2, 4, 6, 8, 14])
ds2 = pd.Series([1, 3, 5, 7, 9])
ds = ds1 + ds2
print("Add two Series:")
print(ds)
print("Subtract two Series:")
ds = ds1 - ds2
print(ds)
print("Multiply two Series:")
ds = ds1 * ds2
print(ds)
print("Divide Series1 by Series2:")
ds = ds1 / ds2
print(ds)
```

```python
#number 5 of series
most_frequent_value = nb['HomeTeamGoals'].value_counts().idxmax()

# Replace values other than the most frequent value with 'Other' in the 'HomeTeamGoals' column
nb['HomeTeamGoals'] = nb['HomeTeamGoals'].where(nb['HomeTeamGoals'] == most_frequent_value, 'Other')

# Print the updated DataFrame
nb
```
